# Python's Lexical Analysis: an interactive demo 

```python
import tokenize
import io
import token

code_bytes = "a + print('sen is cool')"

tokens = tokenize.generate_tokens(io.StringIO(code_bytes).readline)

extracted_tokens = [
    {
        "type": tok.type,
        "type_name": token.tok_name[tok.type],
        "exact_type": tok.exact_type,
        "exact_type_name": token.tok_name[tok.exact_type],
        "string": tok.string,
        "start": tok.start,
        "end": tok.end
    }
    for tok in tokens
]

print(extracted_tokens)
```

```python
class Sentence:

    def __init__(self, sentence:str):
        self.sentence = sentence
        self._index = 0
        self.words = self.sentence.split()

    def __iter__(self):
        return self
    
    def __next__(self):
        if self._index < len(self.words):
            self._index += 1
            return self.words[self._index - 1]
        raise StopIteration
    
    @property
    def index(self):
        return f"The index is {self._index}"
    
    @property
    def chicken(self):
        return f"The index is {self._index}"



sentence = Sentence("Sen is Cool")


print(sentence)
sentence._index = 0

for word in sentence:
    print(word)
```
